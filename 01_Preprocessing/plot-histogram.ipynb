```python
import pandas as pd
import matplotlib.pyplot as plt

# create a pandas dataframe with a column of data
df = pd.DataFrame({'values': [1, 2, 3, 3, 4, 5, 6, 6, 6, 7, 8, 9]})

# plot a histogram of the values column
plt.hist(df['values'], bins=5, color='blue')

# add labels and a title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Values')

# display the plot
plt.show()
```
